```python
# The duration seconds diveristy makes it impossible to categorize, convert the seconds to minutes,then we can analyze the behaviors within certain minutes

data['withinMinutes']=data['durationSeconds'].apply(lambda x: int(x/60)+1)
data.head()

# If the 'withinMinutes'=7, means that this call lasts between 6-7 minutes
```

```python
# Now the categories reduced from 2007 to 74

data['withinMinutes'].value_counts().sort_values(ascending=True)
```

```python
# Convert the date into day, and add an new column called 'day', so we can detect possible pattern whether in a certain day during a week, agents get more calls.
from dateutil import parser

data['day']=data['date'].apply(lambda x:parser.parse(x).strftime("%A"))
data.head()
```
